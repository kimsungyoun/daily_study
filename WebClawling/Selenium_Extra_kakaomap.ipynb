## kakao map 을 이용한 이수맛집 검색

```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import time

from tqdm import tqdm
```

```python
driver = wb.Chrome()
driver.get('https://map.kakao.com/')
driver.maximize_window()
```

```python
click_blue_popup = driver.find_element(By.CSS_SELECTOR, 'body > div.coach_layer.coach_layer_type1 > div > div > div > span')
click_blue_popup.click()
```

```python
search = driver.find_element(By.CSS_SELECTOR, '#search\.keyword\.query')
search.send_keys('이수역 맛집')
search.send_keys(Keys.ENTER)
time.sleep(3)
```

```python
# 객체화
soup = bs(driver.page_source, 'lxml')
```

```python
# 더보기 버튼 누르기
click_more_location = driver.find_element(By.XPATH, '//*[@id="info.search.place.more"]')
click_more_location.click()
time.sleep(2)
```

```python
# 더보기 버튼 누른 후 페이지 번호 
page = driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a')
```

```python
search_page = driver.find_element(By.CSS_SELECTOR,'div.pageWrap')
```

```python
# 추출한 기능들을 통합하여 크롤러 작성
# 예외처리 (try ~ except ~) 
# 성공이든 실패든 '수집완료' 메시지 출력하게 만들기

title_list = []
address_list = []

try:
    while 1:
        index = 0
        for i in tqdm(range(10)):
            #index = 0
            
            if index < 5 :
                page[index].click()
                time.sleep(2)
            
                soup = bs(driver.page_source, 'lxml')
                
                title = soup.select('a.link_name')
                
                for i in title:
                    title_list.append(i.text)
                
                address = soup.select('div.addr > p:nth-child(1)')
                
                for i in address:
                    address_list.append(i.text)
                
                index += 1
            else:
                click_next = driver.find_element(By.XPATH, '//*[@id="info.search.page.next"]')
                click_next.click()
except:
    print('수집 완료')
```
