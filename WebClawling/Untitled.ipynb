```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import pandas as pd

import time

from tqdm import tqdm
```

```python
driver = wb.Chrome()
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%85%B8%EC%9D%B8+%EC%9D%BC%EC%9E%90%EB%A6%AC'
driver.get(url)
driver.maximize_window()
```

```python
page_next = driver.find_element(By.CSS_SELECTOR, 'a.btn_next')
```

```python
title_list = []
content_list = []

try:
    while 1:
        for i in tqdm(range(5)):
            if i < 5:
                page_next = driver.find_element(By.CSS_SELECTOR, 'a.btn_next')
                
                soup = bs(driver.page_source, 'lxml')
                
                title = soup.select('a.news_tit')
                content = soup.select('a.api_txt_lines.dsc_txt_wrap')
                
                for i in title:
                    title_list.append(i.text)
                    
                for i in content:
                    content_list.append(i.text)
                
                page_next.click()
                
            
        break; 
except:
    print('수집 완료')
```

```python
index = 1
for i in title_list:
    print(f'{index} : {i}')
    index += 1
```

```python
index = 1 
for i in content_list:
    print(f'{index} : {i}')
    index += 1
```

```python
driver = wb.Chrome()
driver.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%85%B8%EC%9D%B8+%EC%9D%BC%EC%9E%90%EB%A6%AC')

# 햄버거 버튼 발생
    # 축소 되면서 메뉴 버튼이 바뀌었는데,
    # 창을 최대화해서 진행
    
driver.maximize_window()

titles_list = []
context_list = []

try:
    for i in range(0, 10):
        # 페이지 버튼 재설정
        page = driver.find_elements(By.CSS_SELECTOR, '#main_pack > div.api_sc_page_wrap > div > div > a')
        
        # 페이지 버튼 클릭
        page[i].click()
        time.sleep(2)
        
        # 페이지 소스 가져오기
        page_source = driver.page_source
        
        # BeautifulSoup 객체 생성
        soup = bs(page_source, 'lxml')
        
        # 기사 제목 추출
        titles = soup.select('a.news_tit')
        for title in titles:
            titles_list.append(title.text)
        
        # 뉴스 내용 추출
        contexts = soup.select('div.news_dsc')
        for context in contexts:
            context_list.append(context.text)

except Exception as e:
    print('수집 중 오류가 발생했습니다:', e)
```
