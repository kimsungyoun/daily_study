### Matplotlib
- 데이터 분석 결과를 그래프나 차트로 시각화 해주기 위한 라이브러리
- 분석 결과를 수치만으로 설명하는 것 보다 시각적인 도구들을 활용하면 훨씬 더 직관적으로 이해가 가능하고 타인에게 전달하기도 용이함

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 플롯 차트(plot chart) : 꺾은선 그래프

```python
# x값 없이 선그래프 그리기
y = [2, 4, 6]

# plot : 꺾은 선 그래프 출력 함수
plt.plot(y)
# show : 차트를 화면에 표시해주는 함수
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y)
plt.show()
```

#### 틱(tick) 설정
- 플롯이나 차트에서 축상의 위치 표시 시점을 틱이라고 함
- 이 틱 위에 써진 숫자 혹은 문자는 틱 라벨이라고 함
- 틱 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동설정을 하고싶다면 xticks, yticks 명령을 사용 

```python
x = np.arange(7)
y= [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--',lw = 5, c='c', marker='o', mec='g',mew=5, mfc='r', ms=15)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

#### 그리드(grid)
- 그림의 틱위치를 잘 보여주기 위해 중간에 그리드 선을 그어주는 것

```python
x = np.arange(7)
y= [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--',lw = 5, c='c', marker='o', mec='g',mew=5, mfc='r', ms=15)
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y= [1,4,5,8,9,5,3]

plt.plot(x, y, ls='--',lw = 5, c='c', marker='o', mec='g',mew=5, mfc='r', ms=15)
# 틱이 있는 지점에만 그리드가 생성됨
plt.grid()
plt.show()
```

#### X축, Y축 제목 및 그래프 제목 

```python
x = np.arange(7)
y= [1,4,5,8,9,5,3]


plt.plot(x, y, ls='--',lw = 5, c='c', marker='o', mec='g',mew=5, mfc='r', ms=15)

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title("X-Y graph")
plt.grid()
plt.show()
```

#### 범례 설정 + 여러개 플롯 한꺼번에 그리기

```python
x = np.arange(4)

y1 = [1,2,3,4]
y2 = [2,4,6,8]
y3 = [3,6,9,12]

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

# 범례 설정
# loc : 범례의 위치표시 (upper, center, lower) + (left, right, center)
# prop : 범례 크기 조절
plt.legend(loc='upper right', prop={'size': 7})
plt.show()
```

### 바 차트(bar chart): 막대그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]
```

```python
plt.bar(x, y)
plt.show()
```

```python
# 막대 색상 변경
plt.bar(x, y, color = 'orange')
plt.show()
```

```python
# 원하는 막대 색상 변경
plt.bar(x, y, color = ['orange','cyan','pink','yellow','tomato','lightcoral'])
plt.show()
```

```python
# 막대 테두리 색상 변경
plt.bar(x, y, color = ['orange','tomato','pink'],edgecolor='blue')
plt.show()
```

```python
# 막대 테두리 굵기 변경
plt.bar(x, y, color = ['orange','tomato','pink'],edgecolor='blue',linewidth=4)
plt.show()
```

```python
# 투명도 설정 (0~1 사이의 실수 값,1이면 완전 불투명, 0이면 완전 투명)
plt.bar(x, y, color = ['orange','tomato','pink'],edgecolor='blue',linewidth=4, alpha=0.5)
plt.show()
```

```python
# 막대 너비 설정(디폴트:0.8)
plt.bar(x, y, color = ['orange','tomato','pink'],edgecolor='blue',linewidth=4, alpha=0.5, width = 0.5)
plt.show()
```

### 연습문제
### 사고유형이 차대차 사건 중에 죽거나 다친사람이 많은 발생지시도를 알아보고 시각화하기

- 중요키워드: 차대차, 죽거나 다친사람, 발생지시도

```python
data.columns
```

```python
data['사고유형_대분류'].value_counts()
```

```python
data['사상자수'].value_counts()
```

```python
data['발생지시도'].value_counts()
```

```python
ac = data[data['사고유형_대분류']=='차대차']
```

```python
ac.shape
```

```python
ac[["발생지시도", "사상자수"]]
```

```python
# 1642개의 사건중에서 각 시도의 빈도수 출력
ac['발생지시도'].value_counts()
```

```python
# 각 발생지시도 별로 사상자수의 합계 확인하기
ac2 = ac[['발생지시도','사상자수']].groupby('발생지시도').sum()
ac2
```

```python
x = np.arange(len(ac2.index))
y = ac2["사상자수"].sort_values(ascending=False)

# 그래프 그리기
plt.figure(figsize=(15, 7)) # figure(figsize=(x, y))
plt.bar(x, y)

plt.xticks(x, y.index)
plt.title("차vs차 교통사고의 사상자 수")
plt.xlabel('지역')
plt.ylabel("사상자수")
plt.show()
```
